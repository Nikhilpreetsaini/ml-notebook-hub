# Principal Component Analysis (PCA) Example

This notebook illustrates how to perform Principal Component Analysis (PCA) on the Iris dataset to reduce its dimensionality.
We reduce the four-dimensional features to two principal components and visualize the result.

```
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the first two principal components
plt.figure(figsize=(6, 4))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=iris.target_names[y], palette='Set1', s=70)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.legend(title='Species')
plt.show()

# Show explained variance ratio
print('Explained variance ratio:', pca.explained_variance_ratio_)
```
