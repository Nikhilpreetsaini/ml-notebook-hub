# Polynomial Regression

This notebook shows how to perform polynomial regression by using `PolynomialFeatures` to transform input features and then fitting a linear model.

We create synthetic data and fit a polynomial of degree 2.


```

import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate synthetic data
X, y = make_regression(n_samples=150, n_features=1, noise=20.0, random_state=42)

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

r2 = r2_score(y_test, predictions)
print('R² score (degree 2):', r2)
```
